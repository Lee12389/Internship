```python
import requests as req
from bs4 import BeautifulSoup
import re
import pandas as pd
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os

driver_path = r'D:\Downloads\chrome-win64\chrome-win64\chrome.exe'
os.environ["PATH"] += os.pathsep + driver_path

driver = webdriver.Chrome()
```

```python
def first_question():
    web = req.get('https://www.imdb.com/list/ls056092300/')
    res = BeautifulSoup(web.text)
    movie_list = res.find_all('div',class_='lister-item-content')
    index = 0
    df_movies = pd.DataFrame()
    for lis in movie_list:
        df_movies.loc[index,'name'] = lis.find('h3').find('a').text
        df_movies.loc[index,'year of release'] = int(re.findall('\d+',lis.find('h3').find('span',class_='lister-item-year text-muted unbold').text)[0])
        df_movies.loc[index,'rating'] = float(lis.find('div',class_='ipl-rating-widget').find('span',class_='ipl-rating-star__rating').text)
        index += 1
    return df_movies
```

```python
def second_question():
    
    driver = webdriver.Chrome()
    driver.implicitly_wait(15)
    driver.get('https://peachmode.com/search?q=bags')
    time.sleep(5)
    web = BeautifulSoup(driver.page_source, 'html.parser')
    df = pd.DataFrame({'name':[],'discounted_price':[],'actual_price':[]})
    item_list =web.find('div',class_='st-row').find_all('div',class_='st-product st-col-xs-6 st-col-sm-3 st-col-md-3')
    index = 0
    df = pd.DataFrame({'name':[],'discounted_price':[],'actual_price':[]})
    index = 0
    for item in item_list:
        df.loc[index,'name'] = item.find('div',class_='product-title').find('span').text
        df.loc[index,'discounted_price'] = float(item.find('div',class_='product-prices').find('span',class_='discounted_price st-money money done').text.split(' ')[1].replace(',',''))
        df.loc[index,'actual_price'] = float(item.find('div',class_='product-prices').find('span',class_="price st-money money done").text.split(' ')[1].replace(',',''))
        index += 1
    df['discount_percentage'] = round((df['discounted_price']/df['actual_price'])*100,2)
    driver.quit()
    return df
```

```python
def third_question():
    
    driver = webdriver.Chrome()
    url = 'https://www.icc-cricket.com/'
    driver.get(url)
    time.sleep(5)
    res = BeautifulSoup(driver.page_source,'html.parser')
    uri = [lis for lis in [x.get('href') for x in res.find_all('a')] if 'odi' in lis and '/men' in lis][0]
    driver.get(url[:-1]+uri)
    time.sleep(5)
    res = BeautifulSoup(driver.page_source,'html.parser')
    table = res.find('div',class_="si-table-section").find_all('div',class_="si-table-row")
    df = pd.DataFrame()
    index = 0
    for i in range(1,11):
        df.loc[index,'Position'] = table[i].find('div','si-table-data si-pos').find('span',class_='si-text').text
        df.loc[index,'Team'] = table[i].find('div',class_="si-table-data si-team").find('span',class_="si-fname si-text").text
        df.loc[index,'Matches'] = table[i].find('div',class_="si-table-data si-matches").find('span', class_="si-text").text
        df.loc[index,'Points'] = table[i].find('div',class_='si-table-data si-pts').find('span', class_="si-text").text
        df.loc[index,'Rating'] = table[i].find('div',class_="si-table-data si-rating").find('span', class_="si-text").text
        index += 1
    driver.quit()
    return df
```

```python
def fourth_question():
    
    driver = webdriver.Chrome()
    url = 'https://www.patreon.com/coreyms'
    driver.get(url)
    time.sleep(5)
    res = BeautifulSoup(driver.page_source,'html.parser')
    df_likes = pd.DataFrame()
    index = 0
    for post in res.find_all('div',class_="sc-jvvksu sc-edERGn jUuDJC kpMyHn"):
        df_likes.loc[index,'Heading'] = post.find('span',attrs={'data-tag': 'post-title'}).text
        df_likes.loc[index,'Date'] = post.find('a',attrs={'data-tag': 'post-published-at'}).text
        try:
            df_likes.loc[index,'Content'] = post.find('div',class_="sc-1ye87qi-0 bCBphS").find('p').text
            df_likes.loc[index,'Like_count'] = post.find('button',attrs={'data-tag':"like-button"}).text
        except:
            pass
        index += 1
    driver.quit()
    return df_likes
```

```python
def sixth_question():
    
    driver = webdriver.Chrome()
    url = 'https://www.bewakoof.com/bestseller?sort=popular'
    driver.get(url)
    time.sleep(5)
    res = BeautifulSoup(driver.page_source,'html.parser')
    df_dress = pd.DataFrame()
    index = 0
    for i,pro in enumerate(res.find_all('div',class_="plp-product-card")):
        df_dress.loc[index,'Image_url'] = pro.find('img').get('src')
        df_dress.loc[index,'Price'] = pro.text.split('₹')[1]
        df_dress.loc[index,'Title'] = pro.find('img').get('title')
        index += 1
        if index == 10:
            break
    driver.quit()
    return df_dress
```

```python
def fifth_question():
    
    driver = webdriver.Chrome()
    url = 'https://www.nobroker.in/'
    driver.get(url)
    time.sleep(5)
    button = driver.find_element('xpath',"//*[text()='Buy']")
    button.click()
    loc_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "listPageSearchLocality")))
    loc = ["Jayanagar","Rajaji Nagar","Indira Nagar"]

    for l in loc:
        loc_element.send_keys(l)
        time.sleep(4)
        loc_element.send_keys(Keys.DOWN)
        loc_element.send_keys(Keys.ENTER)
        time.sleep(2)
    search_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//button[contains(@class, 'prop-search-button flex items-center justify-center btn btn-primary btn-lg')]"))
    )
    search_button.click()
    time.sleep(20)
    res = BeautifulSoup(driver.page_source,'html.parser')
    df_house = pd.DataFrame()
    index = 0
    for i,house in enumerate(res.find_all('article')):
        try:
            df_house.loc[index,'Name'] = house.find('a').text
            df_house.loc[index,'Address'] = house.find('div',class_='flex').find('div',class_='mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0.1p po:max-w-95').text    
            df_house.loc[index,'Area'] = house.find('div',id='unitCode').text
            df_house.loc[index,'EMI'] = house.find('div',id='roomType').text
            df_house.loc[index,'Price'] = house.find('div',class_='font-semi-bold heading-6').text
            index += 1
        except:
            pass
    driver.quit()
    return df_house
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def search_properties():
    # Initialize the Chrome webdriver
    driver = webdriver.Firefox()

    try:
        # Open the website
        driver.get('https://www.nobroker.in/')

        # Locations to search
        locations = ['Indira Nagar', 'Jayanagar', 'Rajaji Nagar']

        for location in locations:
            # Find the search box using its name attribute value
            search_box = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.ID, "nbCitySelector"))
            )

            # Enter the location into the search box
            search_box.send_keys(location)

            # Submit the search
            search_box.submit()

            # Wait for page to load
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, "nb__3mmRZ"))
            )

            # Print the current URL
            print(driver.current_url)

            # Clear the search box for the next search
            search_box.clear()

    finally:
        # Close the browser
        driver.quit()

# Call the function
search_properties()
```

```python
class Basefunctions:
    
    def all_col_histplots(self, df):
        plt.figure(figsize=(20,100),facecolor='white')
        plotnum = 1

        for col in df.columns:
            if plotnum < 14:
                ax = plt.subplot(14,1,plotnum)
                sns.histplot(df[col])
                plt.xlabel(col)
            plotnum += 1
        plt.tight_layout()
    
    def scatter(self, df, x, y):
        sns.scatterplot(x=x,y=y,data=df)
        plt.xticks(rotation=90)
        plt.show()
        
    
```

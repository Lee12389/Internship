```python
class EDA:
    
    def __init__(self):
        
        self.target = None
        self.df = None
        
    def all_val_cnts(self):
        for col in self.df.columns:
            print('=+'*30)
            print(self.df[col].value_counts())
            
    def all_col_histplots(self,df = None):
        
        if df == None:
            df = self.df
        plt.figure(figsize=(20,100),facecolor='white')
        plotnum = 1

        for col in df.columns:
            if plotnum < 14:
                ax = plt.subplot(14,1,plotnum)
                sns.histplot(df[col])
                plt.xlabel(col)
            plotnum += 1
        plt.tight_layout()
    def pair_plot(self, df = None):
        
        if df == None:
            df = self.df
        sns.pairplot(self.df)
        
    def get_vif(self, df = None, target = None):
        
        if target == None:
            target = self.target
        if df == None:
            df = self.df
        df = df.drop(columns = [target])
        vif = pd.DataFrame()
        vif['values'] = [variance_inflation_factor(df.values,i) for i in range(len(df.columns))] 
        vif['features'] = df.columns
        print(vif)
    
    def check_skew(self, df = None):
        
        if df == None:
            df = self.df[[col for col in self.df.columns if self.df[col].dtype == 'int' or self.df[col].dtype == 'float']]
        print(self.df.skew())
        
        
```

```python
X = dfx.drop(columns=['Actual_tomorrow_max','Actual_tomorrow_min'])
y = dfx[['Actual_tomorrow_max','Actual_tomorrow_min']]
```

```python
dfx.isnull().sum()
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
X['Date'] = enc.fit_transform(dfx['Date'].values.reshape(-1,1))
```

```python
X.skew()
```

```python
X.drop(columns=['index'],inplace=True)
```

```python
for col,skew in dict(X.skew()).items():
    print(col)
    if skew > 0.5 or skew < 0.5:
        X[col] = np.cbrt(X[col])
        X[col].fillna(np.mean(X[col]),inplace=True)
```

```python
X.skew()
```

        Now we can see a lot of reduction in skewness.It's time for scalar transform

```python
y
```

```python
X.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x = scaler.fit_transform(X)
x = pd.DataFrame(x,columns=X.columns)
```

```python
X_train,X_test, y_train, y_test = train_test_split(x,y['Actual_tomorrow_max'])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor


regressors = {
    'RandomForest': (RandomForestRegressor(), {'n_estimators': [100, 200], 'max_depth': [None, 10, 20]}),
    'GradientBoosting': (GradientBoostingRegressor(), {'n_estimators': [50, 100], 'learning_rate': [0.1, 0.01]}),
    'SVM': (SVR(), {'C': [1, 10], 'kernel': ['linear', 'rbf']}),
    'KNN': (KNeighborsRegressor(), {'n_neighbors':[5,10]}),
    'DecisionTree': (DecisionTreeRegressor(),{'max_depth':[100,200],'max_leaf_nodes':[50,100]})
}
```

```python
bscore = 0
bmodel = ''

for cl, (clas, param_grid) in regressors.items():
    
    grid_search = GridSearchCV(clas, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
    grid_search.fit(X_train,y_train)
    print(grid_search.best_score_)
    if grid_search.best_score_ > bscore:
        bscore = grid_search.best_score_
        bmodel = cl
        params = grid_search.best_params_
print(f"Best model: {bmodel} with accuracy {bscore:.4f}")
```
